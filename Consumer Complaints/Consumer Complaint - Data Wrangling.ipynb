After data exploration, we decided to drop the following columns:

**'Product' and 'Sub-product':** all complaints are relevant to credit card.\
**'Tags':** this column does not covey important meaning. \
**'Sub-issue':** There are already a lot of issues, sub-issue can be dropped.\
**'ZIP code':** 'ZIP code' is highly relevant to 'State', but 'State' conveys message more directly than 'ZIP code'.\
**'Date received':** There are two dates: 'Date received' and 'Date sent to company', in this analysis I will keep only 'Date sent to company', becasue 1. most values of the two columns are identical; 2. slight difference might bed reflected through 'Submitted via'; and 3. 'Date sent to company' is a more accurate measurement for 'Timely response?'\
**'Company public response':** Since around half of the 'Company public response' are missing, and among 68,598 valid values 64,462 are 'Company has responded to the consumer and the CFPB and chooses not to provide a public response'. We decided to delete this column.

```python
# Drop the columns
df_cc = df_cc.drop(['Product', 'Sub-product','Tags', 'Sub-issue', 'ZIP code', 'Date received', 'Company public response'], axis=1)
```
